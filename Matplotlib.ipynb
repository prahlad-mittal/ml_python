```python
#Object Oriented Method for Plotting
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.15,0.5, 0.4,0.4])

axes1.plot(x, x**2)
axes1.set_title('Outer Plot Title', fontsize = 20)
axes1.set_xlabel('xlabel', fontsize = 15, rotation = 20)
axes1.set_ylabel('ylabel', fontsize = 15, rotation = 70)

axes2.plot(x**2, x)
axes2.set_title('Inner Plot Title', fontsize = 17)
axes2.set_xlabel('xlabel', fontsize = 15)
axes2.set_ylabel('ylabel', fontsize = 15)

plt.show()
```

```python
fig, axes = plt.subplots(figsize = (10,6), nrows = 3, ncols = 4)

loc = 0
for i in range(3):
    for j in range(4):
        axes[i][j].plot(x,x**2,'r-', label = 'x vs x**2')
        axes[i][j].plot(x**2,x,'g-', label = 'x**2 vs x')
        if(loc <= 10):
            axes[i][j].legend(loc = loc)
            loc+=1
        else:
            axes[i][j].legend(loc = 0)

plt.tight_layout()
plt.show()
```

```python
# color
# linewidth (lw)
# linestyle (ls)
# marker
# markersize (mc)
# markerfacecolor (mfc)
# markeredgewidth (mew)
# markeredgecolor (mec)

fig, axes = plt.subplots(figsize = (12,6))

axes.plot(x, x+1, color = 'red', linewidth = 1, linestyle = '-', marker = 'o', markersize = 10, markerfacecolor = 'black', markeredgewidth = 3)
axes.plot(x, x+2, color = 'blue', linewidth = 2, linestyle = '-', marker = '+', markersize = 10, markerfacecolor = 'black', markeredgewidth = 1)
axes.plot(x, x+3, color = 'green', linewidth = 1, linestyle = '--', marker = '*', markersize = 10, markerfacecolor = 'black', markeredgewidth = 1)
axes.plot(x, x+4, color = 'purple', linewidth = 2, linestyle = ':', marker = '1', markersize = 10, markerfacecolor = 'black', markeredgewidth = 1)

plt.show()
```

```python
fig, axes = plt.subplots(figsize = (12,6))

axes.plot(x, x+1, color = 'red', lw = 1, ls = '-', marker = 'o', ms = 10, mfc = 'yellow', mew = 3, mec = 'purple')
axes.plot(x, x+2, color = 'blue', lw = 2, ls = '-', marker = '<', ms = 10, mfc = 'yellow', mew = 3, mec = 'red')
axes.plot(x, x+3, color = 'green', lw = 1, ls = '--', marker = 'D', ms = 10, mfc = 'yellow', mew = 3, mec = 'green')
axes.plot(x, x+4, color = 'purple', lw = 2, ls = ':', marker = '>', ms = 10, mfc = 'yellow', mew = 3, mec = 'blue')

plt.show()
```
